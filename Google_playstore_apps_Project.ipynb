### Fixing the Datatype

We can observe that some of datatype of columns are showing like string although they are in numerical and date datatype.
Fixing the data types of columns 'Size', Reviews, installs and price into Numerical datatype

```python
def fix_size(s):
  try:
    if s.endswith('M'):
      return round(float(s[:-1]),2)
    elif s.endswith('k'):
      return round((float(s[:-1]) / 1024),2)
    elif s == '1,000+':
      return 1.0
    else:
      return 0
  except:
      return 0
```

```python
#demo
fix_size('10000 k')
```

```python
# Fixing the Size datatype to Numerical

data['Size'] = data['Size'].apply(fix_size)
data.head()
```

```python
def str_to_num(s):
  try:
    if s.endswith("M"):                 #For Reviews
      return 1000000*int(s[:-1])
    elif s.endswith("k"):
      return 1000*int(s[:-1])
    elif s.endswith("+"):                #For Installs
      s2 = s.replace(',','')
      return int(s2[:-1])
    elif s == "Free":                   # For type
      return 0
    elif s == "Paid":
      return 1
    elif s.startswith("$"):             # For price
      return int(s[1:])
    elif s == "0":
      return 0
    elif s == "Everyone":
      return 0
    else:
      return int(s)

  except:
    return 0
```

```python
data['Reviews'] = data['Reviews'].apply(str_to_num)
data['Installs'] = data['Installs'].apply(str_to_num)
data['Price'] = data['Price'].apply(str_to_num)
data['Type'] = data['Type'].apply(str_to_num)
```
